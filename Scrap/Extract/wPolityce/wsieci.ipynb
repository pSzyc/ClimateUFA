```python
import pandas as pd
import regex as re
from tqdm import tqdm
from pdf_menager import pdf_to_string, pdf_to_txt
from date import date_extractor
from authors import get_authors
from table_of_contents import table_of_contents_extractor
from cleaner import clean_df, weird_cleaner
from text import get_text_by_pages
import os
import seaborn as sns
```

```python
errors = []
results = []
for year in range(2015, 2024):
    for root, dir, files in os.walk(f"wPolityce/{year}/"):
        for file in tqdm(files):
            df, content = table_of_contents_extractor(root + file)
            df = get_text_by_pages(df, content)
            df = get_authors(df, year)
            df['date'] = date_extractor(content, year)
            df['magazine_nr'] = re.search(r"_(\d\d|\d+-\d+|\d)_", file).group(1)
            df = clean_df(df)
            if year==2016:
                df = weird_cleaner(df)
            results.append(df)
```

```python
df_all = pd.concat(results)
df_all.to_csv("wPolityce.csv")
```
